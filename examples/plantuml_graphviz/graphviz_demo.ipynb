```python
import pydot
import cairosvg
# from abbts_blp.toolbox.win_dot_parser import pydot_win
```

```python
# https://graphviz.org/doc/info/shapes.html
dot_string = '''
digraph G{
  rankdir=LR
  node1 [shape=box, label="bbbb"]
  node11 [shape=box, label="bbbsadfb"]
  node2 [shape=diamond]
  node3 [shape=record, label="{ a | {c|abc} | c }"]

  node1 -> node2 [label="bbbb"]
  node11 -> node2 [label="bbbb"]
  node2 -> node3 [label="bbbb"]
}
'''
graphs = pydot.graph_from_dot_data(dot_string)
graph = graphs[0]
graph.write_svg('graphviz_demo.svg')
cairosvg.svg2png(url='graphviz_demo.svg', write_to='graphviz_demo.png', dpi=300)
display.Image('graphviz_demo.png', width=500)
```
